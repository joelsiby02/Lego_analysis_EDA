```python
# i want to merge the 2 dataset using merge fn
merged_df = df.merge(parent_theme, left_on = 'parent_theme', right_on = 'name')
merged_df.head()
```

```python
# on my df there is theme_name & parenyt_name has same info, so delete or drop any one
# i have dropped name_y col from this datsSet
df = merged_df.drop('name_y', axis = 1)
df.head()
```

```python
# Filtering licensed themes from is_licensed col
licensed_df = merged_df[merged_df['is_licensed']]
licensed_df.head()
```

```python
# viewing StarWars licenced theme

# star_wars = licensed_df = [licensed_df['parent_theme'] == 'Star Wars']
# star_wars.head(50)
star_wars = licensed_df[licensed_df['parent_theme'] == 'Star Wars']
star_wars.head(50)
```

```python
# Iterate over 'is_licensed' column and count number of 'True' values
count = 0
for value in star_wars['is_licensed']:
    if value:
        count += 1

# Print number of 'True' values
print('No of licensed StarWars =' ,count)
```

# Q2/ in which year was Star Wars not the most poular licensed theme?

```python
licensed_df.head()
```

```python
licensed_sorted = licensed_df.sort_values('year')
licensed_sorted
```

```python
# Grouping year & parent_theme to make a new col 'count' and store how many were licensed in each year
licensed_sorted.groupby(['year','parent_theme']).head()
```

```python
# Insering a new col into dataSet
licensed_sorted['count'] = 1
licensed_sorted.head()
```

```python
# Now lets count how many times each theme was found in each year and store it in count

sum_df = licensed_sorted.groupby(['year','parent_theme']).sum().reset_index()
sum_df
```

```python
sum_df.groupby(sum_df.index.year).max()
```

```python
sum_df.index = pd.to_datetime(sum_df.index)
sum_df.groupby(sum_df.index.year).max()
```

```python
sum_df = licensed_sorted.groupby(['year','parent_theme']).sum().reset_index() #from above
max_df = sum_df.sort_values('count', ascending = False).drop_duplicates(['year'])
```
